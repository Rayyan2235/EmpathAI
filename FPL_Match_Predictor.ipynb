```python
#Now we will clean data such that we can get float and int64 data do that the ML algo can read it
matches["date"] = pd.to_datetime(matches["date"])
matches.dtypes
```

```python
#Now lets create predictors to add more complexity and make this more functional
matches["venue_code"] = matches["venue"].astype("category").cat.codes
#Converts the venues into a category and then into integers since we can only have 2 inputs "home" and "away"
matches["Opp_code"] = matches["opponent"].astype('category').cat.codes
matches["Hour"] = matches["time"].str.replace(":.+","",regex=True).astype("int") #Removes colons and minutes and keeps only the hour
matches["Day_code"] = matches["date"].dt.dayofweek
# For now we are only trying to find win percentage so L AND D = 0 and W = 1
matches["Target"] = (matches["result"] == "W").astype("int") # returns a boolean using a conditonal, then converted to integer
```

```python
matches
#0 when team is away
#1 when team is home
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=50,min_samples_split=10, random_state=1)
# create training split/data
train = matches[matches["date"] <'2022-01-01']
test = matches[matches["date"] > '2022-01-01']

predictors = ["venue_code", "Opp_code", "Hour","Day_code"] # These are the parameters that we want it to test on for now

#Training the data
rf.fit(train[predictors], train["Target"]) # first param = features to look out for, 2nd = where to train on

#prediction

predicts = rf.predict(test[predictors])
predicts

```

```python
# Now we use metrics to check for accuracy precisiton etc.
```

```python
from sklearn.metrics import accuracy_score # calc accuracy classification score

acc = accuracy_score(test["Target"], predicts) # 1st param = This column holds the real, actual outcomes â€” the true labels. 2nd param where its being compared to
acc
```

```python
#0.61 accuracy

combined =pd.DataFrame(dict(actual=test["Target"], prediction=predicts))
```

```python
# The above is too low, if we say we are more likely to get it wrong when we predict a win
#Therefore we gotta improve this model and this can be done by using rolling averages
```

```python
grouped_matches = matches.groupby(["team"])
```

```python
group = grouped_matches.get_group("Manchester City")
```

```python
group
```

```python
def rolling_averages(group, cols, new_cols):
  group = group.sort_values("date")
  rolling_stats = group[cols].rolling(3, closed='left').mean()
  group[new_cols] = rolling_stats
  group = group.dropna(subset=new_cols) #drop any missing values for rows that havingmissing values, since models cant deal with missing values it will skip it
  return group


```

```python
cols = ["gf", "ga", "sh","sot", "dist","fk", "pk", "pkatt"] # these are the cols we want to compute rolling averages for
new_cols = [f"{c}_rolling" for c in cols]
new_cols
```

```python
# Now lets call the rolling avergaes function
rolling_averages(group,cols,new_cols) # will run only for man city
```

```python
# So if we pass in this for all teams then it will improve accuracy
matches_rolling = matches.groupby("team").apply(lambda x: rolling_averages(x,cols, new_cols))
# take our original matches df, grouped it by team, created 1 df per team then we apply rolling averages to each of these team data frames.
matches_rolling
```
