```python
import pandas as pd
import numpy as np
import kagglehub, os, glob
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score

sns.set_style("whitegrid")
sns.set_palette("RdBu")
```

```python
path = kagglehub.dataset_download("nikhil7280/student-performance-multiple-linear-regression")
print("Downloaded to:", path)
print("Files:", os.listdir(path))

csv_candidates = glob.glob(os.path.join(path, "*.csv"))
assert len(csv_candidates) >= 1, "No CSV found in the downloaded dataset folder."
csv_path = csv_candidates[0]

data = pd.read_csv(csv_path)
display(data.head())
```

```python
display(data.head(5))
```

```python
data.info()
```

```python
print("Dataset shape:", data.shape)
```

```python
null_percent = (data.isnull().sum() / data.shape[0]) * 100
print("Percentage of null values per feature:\n", null_percent)
```

```python
le = LabelEncoder()
data['Extracurricular Activities'] = le.fit_transform(data['Extracurricular Activities'])
```

```python
display(data.head(5))
```

```python
Train = data.drop('Performance Index', axis=1)
Target = data['Performance Index']
```

```python
display(Train.head(5))
```

```python
display(Target.head(5))
```

```python
X_train, X_test, y_train, y_test = train_test_split(Train, Target, test_size=0.2, random_state=42)

print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
print("X_test shape:", X_test.shape)
print("y_test shape:", y_test.shape)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
training_score = model.score(X_train, y_train)
print(f"Training R¬≤ Score: {training_score:.4f}")
```

```python
predict = np.round(model.predict(X_test), 1)

comparison_df = pd.DataFrame({
    "Actual Performance": y_test.values,
    "Predicted Performance": predict
})

print(comparison_df.head(10))
```

```python
hours_studied = 7
previous_score = 99
extra_curricular = 1  # 1 = Yes, 0 = No
sleep_hours = 9
practice_exams_studied = 1

example_student = np.array([[hours_studied, previous_score, extra_curricular, sleep_hours, practice_exams_studied]])
predicted_performance = model.predict(example_student)[0]

print(f"Predicted Performance for sample student: {predicted_performance:.2f}")
```
